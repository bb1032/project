```python
#catch the adv in the list and give every sentence's score
def ScoreF(i):
    if '無法' in i :
        score=-1.75
        deny.append(i)
        return(score)

    elif '沒有' in i :
        score=-1.5
        deny2.append(i)
        return(score)
    elif '不易' in i :
        score=-1.25  
        deny3.append(i)
        return(score)
    elif '非常' in i :
        score=2.25
        good.append(i)
        return(score)
    elif '超' in i :
        score=2
        good1.append(i)
        return(score)
    elif '很' in i :
        score=1.75
        good2.append(i)
        return(score)
    elif '更' in i :
        score=1.5
        good3.append(i)
        return(score)
    elif '小' in i :
        score=1.25
        good4.append(i)
        return(score)
    else:
        score=1
        normal.append(i)
        return(score)
```

```python
#replace the similar word to special word in the content
def replaceF(a,b,c):

    for i in b:
        a=a.replace(i,c)
    return(a)

```
