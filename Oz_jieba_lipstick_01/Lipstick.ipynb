```python
#find the keyword in each product
import json
import jieba.analyse
#jieba.load_userdict('/Users/oz/Desktop/lipstick_dic.txt')


with open('/Users/oz/Desktop/Dior-迪奧_癮誘超模唇膏_750.json') as f:
    test=json.load(f)

with open('/Users/oz/Desktop/stopwords.txt') as f:
    stopwords=f.read().split('\n')
    
dic={}
i=0
while True:
    try:
        test1=test['comments'][i]['content']
    except IndexError :
        break
    words=jieba.lcut(test1)
    for word in words  :
        word=(word.replace('～','')
                  .replace('~','')
                  .replace('↓','')
                  .replace('!','。')
                  .replace('(','')
                  .replace(')','')
                  .replace('\xa0',',')
                  .replace('...',',')
                  .replace(' ）','')
                  .replace('?','')
             )     
      
        if word in stopwords:
            continue
        if word in dic:
            dic[word]+=1
        else :
            dic[word]=0
    i+=1
sorted(dic.items(),key=lambda x : x[1],reverse=True)
    
    
    
```
